```python
import json
import os
import pandas as pd
```

```python
def transform_df(data):
    outp = {
        "pairID": [],
        "input": [],
        "label": [],
        "img_id": []
    }
    for line in data:
        outp['pairID'].append(line['question_id'])
        outp['input'].append(f"prediction: {line['prediction']} explanation: {line['generated_explanation']}")
        outp['label'].append(line['question'])
        outp['img_id'].append(line['img_id'])
    return pd.DataFrame(outp)
```

```python
data_types = ['train', 'dev', 'test']
for d_type in data_types:
    with open(f"gen_{d_type}.json", 'r', encoding='utf-8') as readFile:
        data_ = []
        for line in readFile:
            data_.append(json.loads(line))
    
    with open(f"../../../e-ViL/data/esnlive/esnlive_{d_type}.json", 'r', encoding='utf-8') as readFile:
        o_data_ = []
        for line in readFile:
            o_data_.append(json.loads(line))
    img_id = [o['img_id'] for o in o_data_[0]]        
    data_ = transform_df(data_)
    data_['img_id'] = img_id

    data_.to_csv(f"reverse_{d_type}.tsv", sep='\t', index=False, encoding='utf-8')
```

```python
with open(f"inconsist-variables-test.json", 'r', encoding='utf-8') as readFile:
    variable_ = json.load(readFile)
```

```python
question_id = variable_['pair_id']
```

```python
output = []
for i, que_id in enumerate(question_id):
    pred_ = variable_['context'][i].split(": ")[1].strip()
    d_point_ = {
        "question_id": que_id,
        "label": pred_,
        "sent": variable_['variable'][i],
        "explanation": ['None'],
        "img_id": variable_['img_id'][i],
    }
    output.append(d_point_)
```

```python
output = [output]
```

```python
with open("./esnlive_inconsist_test.json", 'w', encoding='utf-8') as saveFile:
    for line in output:
        json.dump(line, saveFile)
```
